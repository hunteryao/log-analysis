```python
import pandas as pd
import re
import matplotlib.pyplot as plt

filename='E:\Machine learning\data.log'

time=re.compile('20\d{2}-')
time2=re.compile('20\d{2}-\d+-\d+\s\d{2}:\d{2}:\d{2}.\d{3}')

#read file
f=open(filename,'r')
lines=f.read().strip()
f.close()

linelist=re.split(time2,lines)[1:]
times=time2.findall(lines)
element=[]
for line in linelist:
    elements=line.split(' ',4)[1:]
    element.append(elements)


df=pd.DataFrame(element,index=times ,columns=['not sure','loglevel','session','description'])
print (df['loglevel'].unique())
df.groupby('loglevel').size()


    
```

```python
# get the DEBUG dataframe dD
dD=df[df.loglevel=='DEBUG']
l=[]
for Dd in dD['description']:
    l.append(Dd.split(':')[0][:4])
set(l)
```

```python
Dactions=[]
for Dd in dD['description']:
    Dactions.append(Dd.split(':')[0][4:])
dD['action']=Dactions
```

```python
dI=df[df.loglevel=='INFO']
l2=[]
for Di in dI['description']:
    l2.append(Di.split(':')[0][:4])
set(l2)
```

```python
dI=df[df.loglevel=='INFO']
l2=[]
for Di in dI['description']:
    if '[-]' in Di:
        l2.append(Di.split(':')[0][:8])
set(l2)
```

```python
Iactions=[]
dI=df[df.loglevel=='INFO']
for Id in dI['description']:
    if not "[-]" in Id:
        action='REQ'
    elif 'accepted' in Id:
        action='ACCEPT'
    else:action='LINKING'
    Iactions.append(action)
dI['action']=Iactions
```

```python
dW=df[df.loglevel=='WARNING']
Wactions=[]
for Wd in dW['description']:
    #print(Wd)
    if ':' in Wd:
        action=Wd.split(':',1)[0][4:]
    else: action=Wd[4:].strip()
    Wactions.append(action)
#print (len(Wactions))
dW['action']=Wactions

```

```python
print(dI.head(2))
print(dD.head(2))
print(dW.head(2))
```

```python
#vertical merge to dataframe
new_df=pd.concat([dD,dI,dW])

#adjust action column location
cols = list(new_df)
cols.insert(3, cols.pop(cols.index('action')))
new_df =new_df.ix[:, cols]
new_df=new_df.sort_index()

print (new_df.head(2))
new_df.groupby(['loglevel','action']).size()
```

```python
from collections import Counter
import numpy as np
newlabel=[]
for each in (new_df['loglevel'] +'&'+ new_df['action']):
    newlabel.append(each)
counter = Counter(newlabel)
label = counter.keys()
counts = counter.values()

# Plot histogram using matplotlib bar().
indexes = np.arange(len(label))
width =1
plt.bar(indexes, counts, width)
plt.xticks(indexes + width * 0.8, label)
plt.show()
    

```
