```python
import pandas as pd
import re
import matplotlib.pyplot as plt

filename='E:\Machine learning\data.log'

time=re.compile('20\d{2}-')
time2=re.compile('20\d{2}-\d+-\d+\s\d{2}:\d{2}:\d{2}.\d{3}')

#read file
f=open(filename,'r')
lines=f.read().strip()
f.close()

linelist=re.split(time2,lines)[1:]
times=time2.findall(lines)
element=[]
for line in linelist:
    elements=line.split(' ',4)[1:]
    element.append(elements)

#print element
df=pd.DataFrame(element,index=times ,columns=['not sure','loglevel','session','description'])
print (df['loglevel'].unique())
df.groupby('loglevel').size()
```

```python
# get the DEBUG dataframe dD
dD=df[df.loglevel=='DEBUG']
Dactions=[]
for Dd in dD['description']:
    Dactions.append(Dd.split(':')[0][4:])
dD['action']=Dactions
```

```python
dD.head()
```

```python
# DEBUG and REQ description split
dr=dD[dD.action=='REQ']
ldr=[]
for eachline in dr['description']:
    eachline=eachline.split(':',1)[1]
    a=eachline.strip().split('-H',4)[:4]    
    b=eachline.strip().split('-H',4)[4].split('_',1)
    a.extend(b)
    ldr.append(a)
ddr=pd.DataFrame(ldr,dr.index,columns=['a','b','c','d','e','f'])

new_dr=dr.merge(ddr,left_index=True,right_index=True).drop('description',axis=1)
new_dr.head()
```

```python
Iactions=[]
dI=df[df.loglevel=='INFO']
for Id in dI['description']:
    if not "[-]" in Id:
        action='REQ'
    elif 'accepted' in Id:
        action='ACCEPT'
    else:action='LINKING'
    Iactions.append(action)
dI['action']=Iactions
dI.head()
```

```python
#INFO and ACCEPT split
ia=dI[dI.action=='ACCEPT']
lia=[]
for eachline in ia['description']:
    a=eachline.strip().split('accepted')[1][2:-1]
    lia.append(a)
new_ia=ia.copy()

new_ia['IP']=lia
new_ia=new_ia.drop('description',axis=1)
new_ia.head()

```

```python
#INFO and (REQ or LINKING) split
ir=dI[(dI.action=='REQ')|(dI.action=='LINKING')]
match=re.compile('[\[\]]')
lir=[]
for eachline in ir['description']:
    eachline=eachline.strip()
    a=re.split(match,eachline)[1:]
    a[1]=a[1][:-4]
    lir.append(a)
dir=pd.DataFrame(lir,ir.index,columns=['a','b','c','d'])
new_ir=ir.merge(dir,left_index=True,right_index=True).drop('description',axis=1)
new_ir.head()
```

```python
# DEBUG and RESP description split
drs1=dD[dD.action=='RESP']
responses=[]
for eachline in drs1['description']:
    
    #print eachline
    if '[404]' in eachline:
        response=404
        
    elif '[200]' in eachline:
        response=200
    else: response=0
    
    responses.append(response)
set(responses)
drs1['response']=responses
drs1.groupby('response').size()
drs1[drs1['response']==404]
```

```python
drs=drs1[drs1['response']==200]
ldrs=[]
for eachline in drs['description']:
    l=[]
    a= eachline.strip().split('content-length:')[1].strip()
    b=a.split('x-subject-token:')
    l.append(b[0].strip())
    c=b[1].split('vary:')
    l.append(c[0].strip())
    d=c[1].split('RESP BODY:')
    l.append(d[0].strip())
    e= d[1].split('_http_log_response')
    l.append(e[0].strip())
    l.append(e[1])
    ldrs.append(l)

ddrs=pd.DataFrame(ldrs,drs.index,columns=['content-length','x-subject-token','vary','RESP BODY','http_log_response'])

ldrsrb=[]
for each in ddrs['RESP BODY']:
    #print each
    l=[]
    each=each.split('{"methods":')[1]
    a=each.split('"roles":')
    l.append(a[0].strip()[:-1])
    b=a[1].split('"expires_at":')
    l.append(b[0].strip()[:-1])
    c=b[1].split('"project":')
    l.append(c[0].strip()[:-1])
    d=c[1].split('"catalog":')
    l.append(d[0].strip()[:-1])
    e=d[1].split(' "extras":')
    l.append(e[0].strip()[:-1])
    f=e[1].split('"user": ')
    l.append(f[0].strip()[:-1])
    g=f[1].split('"audit_ids": ')
    l.append(g[0].strip()[:-1])
    h=g[1].split('"issued_at":')
    l.append(h[0].strip()[:-1])
    l.append(h[1].strip()[:-2])
    ldrsrb.append(l)
ddrsrb=pd.DataFrame(ldrsrb,drs.index,columns=['methods','roles','expires_at','project','catalog','extras','user','audit_ids','issued_at'])
new_ddrs=ddrs.merge(ddrsrb,left_index=True,right_index=True).drop('RESP BODY',axis=1)
new_drs=drs.merge(new_ddrs,left_index=True,right_index=True).drop('description',axis=1)

new_drs.head()
```
